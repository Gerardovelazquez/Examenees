```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
sns.boxplot(x = df['edad'],
           whis= (0,100), #no mostrar outliers
           ).set_title('Box plot of edad')
plt.show()
```

```python
pivot = pd.pivot_table(df,
                       index = 'desc_ent',
                       values = 'edad',
                        aggfunc = 'median',
                        columns = 'sexo',
                      )
pivot
```
