```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%ls datasets
```

```python
df = pd.read_csv('datasets/antropometria-dataset.csv')
df.head()
```

```python
sns.distplot(df['peso'],
             kde = False,
             bins = 15,
            ).set_title('Histograma de peso', fontsize= 16)
```

```python


sns.lmplot(x='diastol', 
           y='cintura', 
           data = df,
           hue='sexo',
           fit_reg = False, #eliminar linea de regresion
          )
```

```python
sns.boxplot(x = df['cintura'],
           ).set_title('Box plot of cintura')
```

```python
sns.boxplot(x = df['sexo'],
            y = df['cintura'],
           ).set_title('Box plot of sexo_cintura')
plt.show()
```

```python
sns.lmplot(x='peso', 
           y='cintura', 
           data = df,
           hue = 'sexo',
           fit_reg = False, #eliminar linea de regresion
          )
```
